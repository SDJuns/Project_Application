```python
df = pd.read_csv('/content/drive/MyDrive/프로젝트응용(신동준, 이재훈, 최민규)/data/전처리/train_v1.csv')
df = df.drop(columns = ['ROAD_NM_ADDR', 'LOTNO_ADDR', 'X_COORD', 'Y_COORD'])
```

```python
df['MVMN_NM'] = df['MVMN_NM'].fillna('대중교통 등')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# 각 열에 대해 별도의 LabelEncoder 인스턴스 사용
label_gender = LabelEncoder()
df['GENDER'] = label_gender.fit_transform(df['GENDER'])

label_mvmn_nm = LabelEncoder()
df['MVMN_NM'] = label_mvmn_nm.fit_transform(df['MVMN_NM'])

label_mission_check1 = LabelEncoder()
df['TRAVEL_MISSION_CHECK1'] = label_mission_check1.fit_transform(df['TRAVEL_MISSION_CHECK1'])

label_mission_check2 = LabelEncoder()
df['TRAVEL_MISSION_CHECK2'] = label_mission_check2.fit_transform(df['TRAVEL_MISSION_CHECK2'])

label_motive1 = LabelEncoder()
df['TRAVEL_MOTIVE_1'] = label_motive1.fit_transform(df['TRAVEL_MOTIVE_1'])

label_motive2 = LabelEncoder()
df['TRAVEL_MOTIVE_2'] = label_motive2.fit_transform(df['TRAVEL_MOTIVE_2'])

label_motive3 = LabelEncoder()
df['TRAVEL_MOTIVE_3'] = label_motive3.fit_transform(df['TRAVEL_MOTIVE_3'])
```

```python
label_visit = LabelEncoder()
df['VISIT_AREA_NM'] = label_visit.fit_transform(df['VISIT_AREA_NM'])
```

```python
from imblearn.over_sampling import SMOTE

# Separate features and target
X = df.drop('VISIT_AREA_TYPE_CD', axis=1)  # Replace 'TARGET_COLUMN' with actual target column name
y = df['VISIT_AREA_TYPE_CD']  # Replace 'TARGET_COLUMN' with actual target column name

# Apply SMOTE to oversample the minority classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame with the resampled data
df_balanced = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['VISIT_AREA_TYPE_CD'])], axis=1)

# Shuffle the balanced dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the balanced class distribution
print(df_balanced['VISIT_AREA_TYPE_CD'].value_counts())
```
